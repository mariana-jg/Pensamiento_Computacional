```python
def prod_escalar(vector_1, vector_2):
  if len(vector_1) != len(vector_2):
    return None
  producto = 0
  for i in range(len(vector_1)):
    producto += (vector_1[i] * vector_2[i])
  return producto

print(prod_escalar((1,2,3), (1,2,3)))
```

```python
def invertir_lista(lista):
  lista_invertida = []
  for numero in lista:
    lista_invertida.insert(0,numero)
  return lista_invertida

print(invertir_lista([1,3,4,5,2,3]))
```

```python
def obtener_matriz_identidad(tama√±o):
  matriz = []
  for fila in range(tama√±o):
      fil = []
      for columna in range(tama√±o):
          if fila == columna:
              fil.append(1)
          else:
              fil.append(0)
      matriz.append(fil)
  return matriz
```

```python
#Versi√≥n sin filter

def promedio_parcial_intento_1(lista_notas):
        suma_notas = 0
        cantidad_notas = 0
        for nota in lista_notas:
          if nota["Intento"] == 1:
            #suma_notas = suma_notas + nota["Nota"]
            suma_notas += nota["Nota"]
            cantidad_notas += 1 #len(lista_notas) NO
        promedio = suma_notas / cantidad_notas
        return promedio

print(f"El promedio de notas del primer intento del parcial es: {promedio_parcial_intento_1(lista)}")
```

      Par√©ntesis: filter()

      lista_resultante = list(filter(funcion, secuencia))

      Si la funcion devuelve True para un elemento X de la secuencia, ese elemento X "pasa el filtro".

      Par√©ntesis: lambda

      Crea una funci√≥n que no hace falta declararla

      lambda argumentos: expresion

```python
#Versi√≥n con filter

def promedio_parcial_intento_1_filter(lista_notas):
        lista_filtrada = list(filter(lambda nota: nota["Intento"] == 1, lista_notas))
        suma_notas = 0
        for nota in lista_filtrada:
          suma_notas += nota["Nota"]
        promedio = suma_notas / len(lista_filtrada)
        return promedio

print(f"El promedio de notas del primer intento del parcial es: {promedio_parcial_intento_1_filter(lista)}")

#Generalizamos la funci√≥n para que ingrese tambi√©n el intento

def promedio_parcial_filter(lista_notas, intento):
        lista_filtrada = list(filter(lambda nota: nota["Intento"] == intento, lista_notas))
        suma_notas = 0
        for nota in lista_filtrada:
          suma_notas += nota["Nota"]
        promedio = suma_notas / len(lista_filtrada)
        return promedio

print(f"El promedio de notas del primer intento del parcial es: {promedio_parcial_filter(lista, 1)}")
```
